```python
protected = open('/content/entity_group.csv').read().split('\n')[1:-1]
categ = []
for i in range (len(protected)):
  categ.append(protected[i].split(',')[1])
  protected[i] = protected[i].split(',')[0]
 


p = open('/content/prot.txt', encoding ='utf-8-sig').read().split('\n')[:-1]
for i in range (len(p)):
  categ.append(p[i].split(',')[1] )
  p[i] = p[i].split(',')[0] 
  protected.append(p[i])
  

  

print(protected)
print(categ)
```

```python
import re

add_to_dataset=[]
cats = []
for i, term in enumerate(protected):
  try:
    temp = wikipedia.page(term).content.replace('\n','').replace('\t', '').replace('=', '')      
    t = re.split(r'(?<=[^А-Я].[.?]) +(?=[А-Я])', temp)
    add_to_dataset+=t
    for j in range(len(t)):
      cats.append(categ[i])
  except:
    add_to_dataset.append(term)
    cats.append(categ[i])
  continue


add_to_dataset    




```

```python
import pandas as pd

data = pd.DataFrame()
data['message'] = add_to_dataset
data['category'] = cats
data['label'] = False

data.head()
data.to_csv('hui.csv', index = False)
```
